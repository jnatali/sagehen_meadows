```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
profdat = open("prof1.txt", "r")
#reading every row into a list
rows = [(line.strip()).split() for line in profdat]

#remove header row
rows.pop(0)

pos1 = []
elev1 = []
for x in range(len(rows)-1):
    pos1.append(float(rows[x][0]))
    elev1.append(float(rows[x][1]))
    
pos1.reverse() #direction of this survey was opposite to the others, so needs reverse
```

```python
profdat = open("prof2.txt", "r")
#reading every row into a list
rows = [(line.strip()).split() for line in profdat]

#remove header row
rows.pop(0)

pos2 = []
elev2 = []
for x in range(len(rows)-1):
    pos2.append(float(rows[x][0]))
    elev2.append(float(rows[x][1]))
```

```python
profdat = open("prof3.txt", "r")
#reading every row into a list
rows = [(line.strip()).split() for line in profdat]

#remove header row
rows.pop(0)

pos3 = []
elev3 = []
for x in range(len(rows)-1):
    pos3.append(float(rows[x][0]))
    elev3.append(float(rows[x][1]))
```

```python
#mapping approximate depth
elev1b = []
elev2b = []
elev3b = []

depth = 25

for x in elev1:
    elev1b.append(x - depth)
for x in elev2:
    elev2b.append(x - depth)
for x in elev3:
    elev3b.append(x - depth)
```

```python
#elev -- z, pos -- y, separation dist -- x
sep1 = [0]*len(pos1)
sep2 = [-30]*len(pos3)
sep3 = [30]*len(pos2)
```

```python
# plotting ground elevation -- red, given depth below -- blue
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot(sep1,pos1,elev1, 'r')
ax.plot(sep1,pos1,elev1b, 'b')
ax.plot(sep2,pos2,elev2, 'r')
ax.plot(sep2,pos2,elev2b, 'b')
ax.plot(sep3,pos3,elev3, 'r')
ax.plot(sep3,pos3,elev3b, 'b')
ax.set_xlabel('position along the meadow')
ax.set_ylabel('position along cross-section')
ax.set_zlabel('elevation')
```
