```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import datetime
```

```python
#import manual data
manual_readings = pd.read_csv('groundwater_manual_readings.csv')

#find the corresponding compensated level logger file for each row of the manual data
for index, row in manual_readings.iterrows():      #iterate through manual_reading rows
    well_id = row['well_id']
    filename = None
    for f in os.listdir('wells'):                  #iterate through well files
        if not f.startswith('.'):
            if well_id in f:                       #find right file
                filename = f
                break
    if filename == None:
        continue
    logger = pd.read_csv('wells/' + filename, encoding='ISO-8859-1')  #reading the levellogger csv
    logger['timestamp'] = logger['DT']    
    
    #round timestamp to nearest 10 minute interval 
    time = row['timestamp']
    time = pd.to_datetime(time)
    discard = datetime.timedelta(minutes=time.minute % 10, seconds=time.second, microseconds=time.microsecond)
    time -= discard
    if discard >= datetime.timedelta(minutes=5):
        time += datetime.timedelta(minutes=10)
        
    #find levellogger row that corresponds to the correct timestamp
    matching_row = logger.loc[logger['timestamp'] == time]
    print(time)
    print(logger['timestamp'])
    print(matching_row)
    logger_at_time = matching_row.loc[:,'Compensated Level']    #logger level at the correct time
    manual = row['ground_to_water']              #manual level at time, make positive
    manual = manual * -1
        
    logger['delta'] = -1 * (logger['Compensated Level'] - logger_at_time)     #make delta column
    logger['Relative to Ground'] = logger['delta'] + manual                   #make rel to ground column
        
    logger.to_csv('./Relative to Ground/'+ 'rel_to_ground_'+f, encoding='ISO-8859-1')
```
