```python
import numpy as np
import pandas as pd
import os
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.backends.backend_pdf as backend_pdf
```

```python
# Setup directory structure and file names based on structure in github
gw_data_dir = '../../data/field_observations/groundwater/biweekly_manual/'
gw_plot_dir = '../../data/field_observations/groundwater/plots/biweekly_manual/'
gw_old_filename = gw_data_dir + 'groundwater_biweekly_full_2018-2019.csv'
gw_new_filename = gw_data_dir + 'groundwater_biweekly_full_2021.csv'
gw_plot_filename = gw_plot_dir + 'groundwater_biweekly_2018-2021.pdf'

# Setup pdf file for printing and saving all plots
pdf = backend_pdf.PdfPages(gw_plot_filename)
```

```python
# Setup groundwater dataframe from 2018-19 data and 2021 data (weird combo of files right now)
old_gw_data = pd.read_csv(gw_old_filename)
gw_data_2018 = old_gw_data[pd.to_datetime(old_gw_data["timestamp"]).dt.year == 2018]
gw_data_2019 = old_gw_data[pd.to_datetime(old_gw_data["timestamp"]).dt.year == 2019]

gw_data_2021 = pd.read_csv(gw_new_filename)

data = [gw_data_2018, gw_data_2019, gw_data_2021]
data_years = [2018, 2019, 2021]
```

```python
unique_well_names = gw_data_2021["well_id"].unique()
print(len(unique_well_names))
```

```python
for well_id in unique_well_names:
    
    print(well_id)
    fig, axis = plt.subplots(figsize=(20, 10), nrows=1, sharex=True)
    
    # loop through gw dataframes by year
    for i in range(len(data)):
        gw_data = data[i]
        well_data = gw_data[gw_data["well_id"] == well_id]
        
        # FIXED in data file on 04/26/2022: For the 2018 and 2019 data, the column is named differently, and the distances are negative
        #if i < 2:
        #    groundwater_data = well_data["ground_to_water_m"]
        #else:
        #    groundwater_data = -1*well_data["ground_to_water_cm"]
        
        groundwater_data = well_data["ground_to_water_cm"]
        
        # Set year to a common year for all data so they plot on top of each otherr
        time_data = pd.to_datetime(well_data["timestamp"]).apply(lambda x: x.replace(year=2000))
        
        # Plot as line with point markesr for each data point
        axis.plot(time_data, groundwater_data, label=data_years[i])
        axis.scatter(time_data, groundwater_data, label=data_years[i])
    
    axis.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
    axis.set_xlabel("Date", size=20)
    axis.set_ylabel("Ground to Water (cm)", size=20)
    plt.title("Groundwater data for well " + well_id, size=20)
    plt.legend()
    plt.show()
    pdf.savefig(fig)
    
pdf.close()
```
