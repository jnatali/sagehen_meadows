```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Example of raw logger data

```python
logger = pd.read_csv('data/EEF-1_2018_0717_0725.csv', header=11, encoding='ISO-8859-1')
logger['DT'] = pd.to_datetime(logger['Date'] + ' ' + logger['Time'])

fig, (ax1, ax2) = plt.subplots(figsize=(20, 10), nrows=2, sharex=True)
ax1.plot(logger['DT'], logger['LEVEL'])
ax1.set_ylabel('Level', size=20)
ax2.plot(logger['DT'], logger['TEMPERATURE'])
ax2.set_ylabel('Temperature', size=20)

for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(13)
for tick in ax2.yaxis.get_major_ticks():
    tick.label.set_fontsize(13)
for tick in ax2.xaxis.get_major_ticks():
    tick.label.set_fontsize(13)

fig.suptitle('Level and Temperature', size=25)
```

## Example of cutting that same logger

```python
logger = pd.read_csv('data/EEF-1_2018_0717_0725.csv', header=11, encoding='ISO-8859-1')
logger['DT'] = pd.to_datetime(logger['Date'] + ' ' + logger['Time'])

# Calculate rate of change
logger['Rate of Change'] = np.append(np.abs((logger['TEMPERATURE'].iloc[:-1].values - logger['TEMPERATURE'].iloc[1:].values)/(10)), 0.0001)

# Split data in half
first = logger.iloc[:int(logger.shape[0]/2)]
second = logger.iloc[int(logger.shape[0]/2):]

# In first half, cut out all data before the last point that has a rate of change greater than 0.001
index = first.loc[(first['Rate of Change'] >= 0.001), :].index[-1]
first_cut = first.iloc[index:]

# In second half, cut out all data after the first point that has a rate of change greater than 0.001
index = second.loc[(second['Rate of Change'] >= 0.001), :].index[0]
second_cut = second.loc[(second.index < index)]

# Snip both halfs back together
logger = pd.concat((first_cut, second_cut), axis=0)

print('# of rows cut off of front end:', first.shape[0] - first_cut.shape[0])
print('# of rows cut off of back end:', second.shape[0] - second_cut.shape[0])

fig, (ax1, ax2) = plt.subplots(figsize=(20, 10), nrows=2, sharex=True)
ax1.plot(logger['DT'], logger['LEVEL'])
ax1.set_ylabel('Level', size=20)
ax2.plot(logger['DT'], logger['TEMPERATURE'])
ax2.set_ylabel('Temperature', size=20)

for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(13)
for tick in ax2.yaxis.get_major_ticks():
    tick.label.set_fontsize(13)
for tick in ax2.xaxis.get_major_ticks():
    tick.label.set_fontsize(13)

fig.suptitle('Level and Temperature', size=25)
```

## Cutting data for all loggers

```python
for f in os.listdir('data'):
    if not f.startswith('.'):
        logger = pd.read_csv('data/' + f, header=10, encoding='ISO-8859-1')
        if 'Date' in logger:
            logger['DT'] = pd.to_datetime(logger['Date'] + ' ' + logger['Time'])
        else:
            logger = pd.read_csv('data/' + f, header=11, encoding='ISO-8859-1')
            logger['DT'] = pd.to_datetime(logger['Date'] + ' ' + logger['Time'])
        
        print('File name:', f)
        
        # Calculate rate of change
        logger['Rate of Change'] = np.append(np.abs((logger['TEMPERATURE'].iloc[:-1].values - logger['TEMPERATURE'].iloc[1:].values)/(10)), 0.0001)

        # Split data in half
        first = logger.iloc[:int(logger.shape[0]/2)]
        second = logger.iloc[int(logger.shape[0]/2):]

        # In first half, cut out all data before the last point that has a rate of change greater than 0.001
        try:
            index = first.loc[(first['Rate of Change'] >= 0.001), :].index[-1]
            first_cut = first.iloc[index:]
        except IndexError as err:
            first_cut = first
            print('Nothing to cut off front end.')

        # In second half, cut out all data after the first point that has a rate of change greater than 0.001
        try:
            index = second.loc[(second['Rate of Change'] >= 0.001), :].index[0]
            second_cut = second.loc[(second.index < index)]
        except IndexError as err:
            second_cut = second
            print('Nothing to cut off back end.')
        

        # Snip both halfs back together
        logger = pd.concat((first_cut, second_cut), axis=0)

        print('# of rows cut off of front end:', first.shape[0] - first_cut.shape[0])
        print('# of rows cut off of back end:', second.shape[0] - second_cut.shape[0])

        fig, (ax1, ax2) = plt.subplots(figsize=(20, 10), nrows=2, sharex=True)
        ax1.plot(logger['DT'], logger['LEVEL'])
        ax1.set_ylabel('Level', size=20)
        ax2.plot(logger['DT'], logger['TEMPERATURE'])
        ax2.set_ylabel('Temperature', size=20)

        for tick in ax1.yaxis.get_major_ticks():
            tick.label.set_fontsize(13)
        for tick in ax2.yaxis.get_major_ticks():
            tick.label.set_fontsize(13)
        for tick in ax2.xaxis.get_major_ticks():
            tick.label.set_fontsize(13)

        fig.suptitle('Level and Temperature', size=25)
        plt.show()
```
