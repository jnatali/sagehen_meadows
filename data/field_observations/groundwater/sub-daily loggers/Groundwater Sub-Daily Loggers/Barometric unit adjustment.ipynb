```python
import pandas as pd
import numpy as np
import os
from os import chdir
from glob import glob
```

```python
bar_file_path = "./baro_data/"
chdir(bar_file_path)
list_of_files = [file for file in glob("*")]
```

```python
#combining all of the original barometric csvs into one

def produceOneDataframe(list_of_files):
   # Consolidate all CSV files into one object
   all_baro_data = pd.concat([pd.read_csv(file, header=10, encoding='ISO-8859-1') for file in list_of_files])
   return all_baro_data
    
baro_data_in_kPa = produceOneDataframe(list_of_files)
```

```python
# converting the 'level' column from kPa to meters

baro_data_in_kPa['LEVEL'] *= 0.101972
baro_data_in_kPa
```
