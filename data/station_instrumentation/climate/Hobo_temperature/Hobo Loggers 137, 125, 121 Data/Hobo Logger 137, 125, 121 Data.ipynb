### Importing, Merging, Cleaning, and Saving All Data for Logger #137

```python
# Importing all data
csv1 = pd.read_csv('data/SGH_HOBO_137_11:22:16-4:4:17.csv', skiprows=1, index_col='#')
csv2 = pd.read_csv('data/SGH_HOBO_137_11:3:17-4:9:18.csv', skiprows=1, index_col='#')
csv3 = pd.read_csv('data/SGH_HOBO_137_12:13:18-4:15:19.csv', skiprows=1, index_col='#')
csv4 = pd.read_csv('data/SGH_HOBO_137_2:13:20-9:25:20.csv', skiprows=1, index_col='#')
csv5 = pd.read_csv('data/SGH_HOBO_137_4:15:19-9:27:19.csv', skiprows=1, index_col='#')
csv6 = pd.read_csv('data/SGH_HOBO_137_4:4:17-8:18:17.csv', skiprows=1, index_col='#')
csv7 = pd.read_csv('data/SGH_HOBO_137_4:9:18-7:2:18.csv', skiprows=1, index_col='#')
csv8 = pd.read_csv('data/SGH_HOBO_137_7:2:18-8:14:18.csv', skiprows=1, index_col='#')
csv9 = pd.read_csv('data/SGH_HOBO_137_8:18:17-11:3:17.csv', skiprows=1, index_col='#')
csv10 = pd.read_csv('data/SGH_HOBO_137_9:27:19-2:13:20.csv', skiprows=1, index_col='#')
csv11 = pd.read_csv('data/SGH_HOBO_137_8:14:18-12:13:18.csv', skiprows=1, index_col='#')

# Merging all data
hobo_137 = pd.concat([csv1, csv2, csv3, csv4, csv5, csv6, csv7, csv8, csv9, csv10, csv11])

# Renaming columns to be more friendly
hobo_137.rename({'Date Time, GMT-07:00': 'DT',
                'Temp, °F (LGR S/N: 10769296, SEN S/N: 10769296, LBL: sgh 137 temp)': 'Temperature (F)',
                'RH, % (LGR S/N: 10769296, SEN S/N: 10769296, LBL: sgh 137 rh)': 'RH (%)',
                'Coupler Detached (LGR S/N: 10769296)': 'Coupler Detached',
                'Coupler Attached (LGR S/N: 10769296)': 'Coupler Attached',
                'Stopped (LGR S/N: 10769296)': 'Stopped',
                'End Of File (LGR S/N: 10769296)': 'End of File',
                'Bad Battery (LGR S/N: 10769296)': 'Bad Battery',
                'Good Battery (LGR S/N: 10769296)': 'Good Battery'},
                axis=1, inplace=True)

# Converting Date column to Datetime Object for easier manipulation
hobo_137['DT'] = pd.to_datetime(hobo_137['DT'])

# Sorting by Date column
hobo_137.sort_values(by='DT', inplace=True)

# Saving to csv
hobo_137.to_csv('data/137_All.csv')

hobo_137.head()
```
