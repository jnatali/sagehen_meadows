```python
#if the precipitation has a negative value, make it zero
climate_data.loc[(climate_data.Daily_Precipitation < 0), 'Daily_Precipitation'] = 0

#make it daily precipitation instead of cumulative
precip_difference = climate_data.Daily_Precipitation.diff()
climate_data['Precip_Difference'] = precip_difference

for i in np.arange(climate_data.shape[0]):
    current_precip = climate_data['Daily_Precipitation'][i]
    
    #subtract previous day from current day to get current day's precipitation
    if climate_data['Precip_Difference'][i] >= 0:
        current_precip = climate_data['Precip_Difference'][i]
    else: 
        current_precip = 0
    
    climate_data['Daily_Precipitation'][i] =  current_precip
climate_data.reset_index(inplace=True)
climate_data.rename(columns={"index": "timestamp"}, inplace=True)
climate_data.fillna(0)
```

```python
#average groundwater data by day

cols = gw_data.columns.difference(['timestamp'])

#got this from stack overflow
gw_data[cols] = gw_data[cols].astype(float)
gw_data = gw_data.resample('d', on='timestamp').mean().dropna(how='all')
gw_data.rename(columns={'ground_to_water': 'avg ground_to_water'}, inplace=True)
gw_data.reset_index(inplace=True)
gw_data.rename(columns={"index": "timestamp"}, inplace=True)
gw_data
```

```python
#take 2-day average

i = 0
drop_indices = []
while i < gw_data.shape[0] - 1:
    if gw_data['timestamp'][i+1] == gw_data['timestamp'][i] + np.timedelta64(1, 'D'): 
        two_day_avg = (gw_data['avg ground_to_water'][i] + gw_data['avg ground_to_water'][i+1])/2
        gw_data['avg ground_to_water'][i+1] = two_day_avg
        drop_indices.append(i)
        i += 1
    else:
        i += 1
#reindex
gw_data = gw_data.drop(gw_data.index[drop_indices])
gw_data.reset_index(inplace=True)
gw_data = gw_data.drop('index', axis=1)
gw_data
```

```python
#take 2-day average

i = 0
drop_indices = []
while i < gw_data.shape[0] - 1:
    if gw_data['timestamp'][i+1] == gw_data['timestamp'][i] + np.timedelta64(1, 'D'): 
        count = 0
        running_sum = gw_data['avg ground_to_water'][i]
        while (i + count < gw_data.shape[0] - 1) and (gw_data['timestamp'][i + count + 1] == gw_data['timestamp'][i + count] + np.timedelta64(1, 'D')): 
            running_sum += gw_data['avg ground_to_water'][i+count + 1]
            drop_indices.append(i + count)
            count += 1
        gw_data['avg ground_to_water'][i+count] = running_sum/(count + 1)
        i += (count + 1)
    else:
        i += 1
#reindex
gw_data = gw_data.drop(gw_data.index[drop_indices])
gw_data.reset_index(inplace=True)
gw_data = gw_data.drop('index', axis=1)
gw_data
```

```python
#combine the groundwater and climate data into one dataframe

combined_data = climate_data.merge(gw_data, on=['timestamp'], how='left')
pd.set_option('display.max_rows', combined_data.shape[0]+1)
combined_data
```

```python
#PLOT
fig, (ax1, ax2, ax3) = plt.subplots(figsize=(20, 10), nrows=3, sharex=True)
ax1.plot(combined_data['timestamp'], combined_data['Daily Max Air Temp'])
ax1.set_ylabel('Max Temp, C', size=20)
ax2.bar(combined_data['timestamp'], combined_data['Daily_Precipitation'])
ax2.set_ylabel('Precipitation, cm', size=20)
ax3.bar(combined_data['timestamp'], combined_data['avg ground_to_water'])
ax3.set_ylabel('Ground to Water, cm', size=20)

for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(13)
for tick in ax2.yaxis.get_major_ticks():
    tick.label.set_fontsize(13)
for tick in ax2.xaxis.get_major_ticks():
    tick.label.set_fontsize(13)

fig.suptitle('Groundwater Level and Climate', size=25)
```
